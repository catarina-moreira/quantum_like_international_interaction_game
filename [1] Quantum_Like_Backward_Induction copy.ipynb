```python
# Full code including the function and its application

def find_matching_theta(data):
    """
    Find the values of theta1 and theta2 where MAX_Quantum equals groundtruth, along with their count.

    :param data: DataFrame containing the data.
    :return: DataFrame with filtered rows where MAX_Quantum equals groundtruth and count of each combination.
    """
    # chanmge the values of column groundtruth to uppercase
    data['groundtruth'] = data['groundtruth'].str.upper()
    data['MAX_Quantum'] = data['MAX_Quantum'].str.upper()
    
    
    matching_rows = data[data['MAX_Quantum'] == data['groundtruth']]
    
    # Counting the occurrences of each combination of theta1 and theta2
    matching_counts = matching_rows[['theta1', 'theta2']].groupby(['theta1', 'theta2']).size().reset_index(name='count')
    return matching_counts

# Function to find the top 10 peaks in the data
def find_top_peaks(data, num_peaks=10):
    flat_indices = np.argpartition(-data.ravel(), num_peaks)[:num_peaks]
    top_peaks =  [np.unravel_index(idx, data.shape) for idx in flat_indices]
    top_peaks = np.sort(top_peaks)[-1*num_peaks:]
    return top_peaks

# Get the indices of the top 10 peaks
# top_peaks_indices = find_top_peaks(values, num_peaks=10)

# Extract the theta1, theta2, and value for each top peak
# top_peaks_theta1 = [theta1[idx[0]] for idx in top_peaks_indices]
# top_peaks_theta2 = [theta2[idx[1]] for idx in top_peaks_indices]
# top_peaks_values = [values[idx] for idx in top_peaks_indices]
```

```python
def generate_contour_plot( thetas_df, title, xaxis_title, yaxis_title, width=700, height=700, margin=dict(l=65, r=50, b=65, t=90) ):
    
    # Splitting the 'key' column by "_" and creating new columns theta1 and theta2
    thetas_df[['theta1', 'theta2']] = thetas_df['key'].str.split('_', expand=True).astype(float)

    # Pivot the data to create a grid of theta1, theta2, and count
    pivot_data = thetas_df.pivot("theta1", "theta2", "count")
    theta1 = pivot_data.index
    theta2 = pivot_data.columns
    values = pivot_data.values

    # Handling missing values, if any, by filling with zeros
    values = np.nan_to_num(values)
    
    # Define the minimum and maximum values for the color scale
    zmin = 0
    zmax = 121

    # Create the contour trace with zmin and zmax specified
    contour_trace = go.Contour(z=values, x=theta1, y=theta2, colorscale='Viridis', zmin=zmin, zmax=zmax)

    fig = go.Figure(data=[contour_trace])
    
    
    #fig = go.Figure(data=[go.Contour(z=values, x=theta2, y=theta1, colorscale='Viridis')])
    fig.update_layout(title=title,
                      xaxis_title=xaxis_title,
                      yaxis_title=yaxis_title,
                      autosize=False,
                      width=width,
                      height=height,
                      margin=margin)
    
    # update limits of the x-axis to -2pi to 2pi
    fig.update_xaxes(range=[theta1.min(), theta1.max()])
    fig.update_yaxes(range=[theta2.min(), theta2.max()])
    
    # add axis labels
    fig.update_xaxes( title_text="Theta1" )
    fig.update_yaxes( title_text="Theta2" )
    
    # find the top 20 peaks
    top_peaks_indices = find_top_peaks(values, num_peaks=10)
    
    # extract the theta1, theta2, and value for each top peak
    top_peaks_theta1 = [theta1[idx[0]] for idx in top_peaks_indices]
    top_peaks_theta2 = [theta2[idx[1]] for idx in top_peaks_indices]
    top_peaks_values = [values[idx] for idx in top_peaks_indices]
    
    # add the top 20 peaks to the plot and add the hover thetas
    fig.add_trace(go.Scatter(x=top_peaks_theta1, y=top_peaks_theta2, mode='markers', marker=dict(color='red', size=10)))
    fig.update_traces(hovertemplate='Theta1: %{x}<br>Theta2: %{y}')
    
    
    # find 4 clusters of top_peaks_theta1 and top_peaks_theta2
    X = np.array(list(zip(top_peaks_theta1, top_peaks_theta2)))
    
    kmeans = KMeans(n_clusters=4, random_state=0).fit(X)
    labels = kmeans.labels_
    
    # add the cluster labels to the dataframe
    top_peaks_df = pd.DataFrame(list(zip(top_peaks_theta1, top_peaks_theta2, labels)), columns=['theta1', 'theta2', 'cluster'])
    
    # add the cluster labels to the plot
    #fig.add_trace(go.Scatter(x=top_peaks_df['theta2'], y=top_peaks_df['theta1'], mode='markers', marker=dict(color=labels, size=10, colorscale='Magma')))
    #fig.update_traces(hovertemplate='Theta1: %{y}<br>Theta2: %{x}<br>Cluster: %{marker.color}')
    
    # add a red rectangle around EACH cluster
    
    
    cluster_colors = ['#C0392B',  '#3498DB', '#E67E22','#F4D03F']
    for i in range(0, 4):
        cluster = top_peaks_df[top_peaks_df['cluster'] == i]
        fig.add_shape(type="rect",
            x0=cluster['theta1'].min()-0.1, y0=cluster['theta2'].min()-0.1, x1=cluster['theta1'].max()+0.1, y1=cluster['theta2'].max()+0.1,
            line=dict(color=cluster_colors[i],width=3))
        fig.add_trace(go.Scatter(x=cluster['theta1'], y=cluster['theta2'], mode='markers', marker=dict(color=cluster_colors[i], size=10)))
        fig.update_traces(hovertemplate='Theta1: %{x}<br>Theta2: %{y}')
        
        # add a text label to each cluster

        if (i == 0):
            fig.add_annotation(
                x=cluster['theta1'].mean() ,
                y=cluster['theta2'].mean()-0.8,
                text="Cluster " + str(i+1),
                showarrow=False,
                font=dict(
                    family="Courier New, monospace",
                    size=16,
                    color="#ffffff"
                    ),
                align="center",
                bordercolor="#c7c7c7",
                borderwidth=2,
                borderpad=4,
                bgcolor=cluster_colors[i],
                opacity=0.8
                )
        
        if (i == 1):
            fig.add_annotation(
                x=cluster['theta1'].mean() ,
                y=cluster['theta2'].mean() + 0.6,
                text="Cluster " + str(i+1),
                showarrow=False,
                font=dict(
                    family="Courier New, monospace",
                    size=16,
                    color="#ffffff"
                    ),
                align="center",
                bordercolor="#c7c7c7",
                borderwidth=2,
                borderpad=4,
                bgcolor=cluster_colors[i],
                opacity=0.8
                )
        
    # remove the traces from the legend
    fig.update_layout(showlegend=False)
    
    fig.show()
    
    return top_peaks_df
    
top_peaks_df = generate_contour_plot( thetas_df, 'Interactive Contour Plot of Theta Values and Count', 'Theta2', 'Theta1', width=700, height=700, margin=dict(l=65, r=50, b=65, t=90) )
```
